```python
import pandas as pd
import numpy as np
```

# Data preparation

```python
# Load the model
df = pd.read_csv('data.csv')
# Take a look at the first rows to see what are we working with
df.head(n=5)
```

```python
# make columns consistent (lowercase and replace whitespaces with underscore)
# We can get the columns (It's an index)
df.columns
```

```python
# We can run string operations like lower() and replace(" ", "_")
df.columns.str.lower().str.replace(" ", "_")
```

```python
# So we run lower() and replace(" ", "_") and assign it back to df.columns
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head(n = 5)
```

```python
# We can check the types of each column
# We are looking for the strings... in this case the type object are the strings
# Because we are loading a csv we can be pretty sure that the objects are strings
df.dtypes
```

```python
# df.dtypes[df.dtypes == 'object'].index
# We can get an Index of the columns with the dtype of object
df.columns[df.dtypes == 'object']
```

```python
# And transform that Index in a Python list
string_columns = list(df.columns[df.dtypes == 'object'])
string_columns
```

```python
# Then loop through that list running lower and replace(" ", "_") on each of the columns
for col in string_columns:
    df[col] = df[col].str.lower().str.replace(" ", "_")
df.head(n = 5)
```
